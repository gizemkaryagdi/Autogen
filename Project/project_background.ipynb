```python
import autogen
from autogen.agentchat.groupchat import GroupChat,GroupChatManager
from autogen.coding import LocalCommandLineCodeExecutor
from autogen.agentchat.user_proxy_agent import UserProxyAgent
from autogen.agentchat.assistant_agent import AssistantAgent
```

```python
config_list = [ 
    {
        "model": "llama3",
        "base_url": "http://localhost:11434/v1",
        "api_key": "ollama",
    }
]
```

```python
llm_config={
    "cache_seed":41,
    "temperature":0,
    "config_list":config_list,
    "timeout":120,
}
```

```python
user_proxy=autogen.UserProxyAgent(
    name="Admin",
    system_message="A human admin. Interact with the planner to discuss the plan. Plan execution needs to be approved "
                   "by this admin.",   
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
    code_execution_config={
        "work_dir": "code",
        "use_docker": False,
    },
    human_input_mode="TERMINATE",       
)
```

```python
planner = autogen.AssistantAgent(
    name="Planner",
    system_message="""Planner. Suggest a plan.""",
    llm_config=llm_config,
)
```

```python
engineer = autogen.AssistantAgent(
    name="Engineer",
    llm_config=llm_config,
    system_message="""Engineer.You are working as a Senior Frontend Engineer. Write your only thing in CSS code. For that, visit this website : https://webaim.org/resources/contrastchecker/""",
)
```

```python
critic = autogen.AssistantAgent(
    name="Critic",
    system_message="Critic. Double check plan provide feedback.",
    llm_config=llm_config,
)
```

```python
groupchat = GroupChat(agents=[user_proxy, planner,engineer, critic], messages=[], max_round=12, speaker_selection_method="round_robin")
manager = GroupChatManager(groupchat=groupchat, llm_config={"config_list": config_list})
```

```python
user_proxy.initiate_chat(
    manager,
    message="""I'm getting an error.The error "Background and foreground colors do not have sufficient contrast ratio." '<<span style="color:#58E7E7;">>' It means that the color contrast between the text and the background on a web page or application is not sufficient. This error is especially common in designs that do not meet accessibility standards. Insufficient contrast can make text difficult or impossible to read, especially for users with visual impairments.
""",
)
```
